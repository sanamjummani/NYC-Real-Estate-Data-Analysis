RUNNING REPLACEMENT OF CHARACTERS AND BLANKS TO "NAN" AGAIN TO REFLECT ON NUMERICAL COLUMNS NOW:

```python
df['GROSS SQUARE FEET'] = df['GROSS SQUARE FEET'].replace({0:np.nan})
```

REMOVING OUTLIERS IN SALE PRICE COLUMN USING IQR METHOD: 

```python
q1 = df['SALE PRICE'].quantile(0.25)
q1
```

```python
q3 = df['SALE PRICE'].quantile(0.75)
q3
```

```python
iqr = q3-q1
iqr
```

```python
upper_bound = q3+1.5*iqr
upper_bound
```

```python
lower_bound = q1-1.5*iqr
lower_bound
```

```python
df[df['SALE PRICE'] < 0]
```

```python
df_outliers = df[(df['SALE PRICE'] < lower_bound) | (df['SALE PRICE'] > upper_bound)]
df_outliers.shape
df_outliers
```

```python
df_without_outliers = df[(df['SALE PRICE'] > lower_bound) & (df['SALE PRICE'] < upper_bound)]
df_without_outliers
```
